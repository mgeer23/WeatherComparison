```python
import json, pprint, datetime
from os.path import exists
```

First extract data for the past 24 hours from met_past_24h.jsonl

```python
with open('data/met_past_24h.jsonl', 'r') as f:
    line = json.loads(f.readline())

pprint.pp(line)
print(type(line))
```

```python
parameters_key = line['SiteRep']['Wx']['Param']
parameters_key
```

```python
pprint.pp(line['SiteRep']['DV'])
location = line['SiteRep']['DV']['Location']['name']
print('location is ' + location)
```

```python
days = line['SiteRep']['DV']['Location']['Period']

for day in days:
    print(day)

pprint.pp(day)
type(day)
```

Weather code definitions link:
https://www.metoffice.gov.uk/services/data/datapoint/code-definitions

```python
date = day['value'][:-1]
for h in day['Rep']:
    temp = h['T']
    weather_code = h['W']
    precip_bool = int(int(weather_code) > 8)
    hour = int(int(h['$']) / 60)
    

pprint.pp(h)
print(date)
print(temp)
print(weather_code)
print(precip_bool)
print(hour)
```

```python
print(str.split(date, '-'))
split_date = str.split(date, '-')
dt = datetime.datetime(int(split_date[0]), int(split_date[1]), int(split_date[2]), hour)
print(dt)

print(location)
print(date)
print(hour)
print(temp)
print(weather_code)

data_row = [location, dt, temp, weather_code, precip_bool]
data_row
```

```python
if exists('data/met_past_24h.csv'):
    with open('data/met_past_24h.csv', 'a') as f:
        f.write(str(data_row)[1:-1] + '\n')
else:
    with open('data/met_past_24h.csv', 'a') as f:
        f.write('location, date, hour, temp, weather_code, was_raining' + '\n')
        f.write(str(data_row)[1:-1] + '\n')
```

MariaDB met_past_24h table creation:

CREATE TABLE met_past_24h(
    location VARCHAR(255),
    date_time DATETIME,
    temp FLOAT,
    weather_code SMALLINT,
    precip_bool BOOL,
    PRIMARY KEY(location, date_time)
);
