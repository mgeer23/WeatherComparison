First extract data for the past 24 hours from met_past_24h.jsonl

```python
with open('data/met_past_24h.jsonl', 'r') as f:
    line = json.loads(f.readline())

pprint.pp(line)
print(type(line))
```

```python
pprint.pp(line['SiteRep']['DV'])
location = line['SiteRep']['DV']['Location']['name']
print('location is ' + location)
```

```python
days = line['SiteRep']['DV']['Location']['Period']

for day in days:
    print(day)

pprint.pp(day)
type(day)
```

Weather code definitions link:
https://www.metoffice.gov.uk/services/data/datapoint/code-definitions

```python
date = day['value'][:-1]
for h in day['Rep']:
    temp = h['T']
    weather_code = h['W']
    precip_bool = int(int(weather_code) > 8)
    hour = int(int(h['$']) / 60)
    

pprint.pp(h)
print(date)
print(temp)
print(weather_code)
print(precip_bool)
print(hour)
```

```python
print(str.split(date, '-'))
split_date = str.split(date, '-')
dt = datetime.datetime(int(split_date[0]), int(split_date[1]), int(split_date[2]), hour)
print(dt)

stringtime = f"{date} {hour}:00:00"
print('string time: ' + stringtime)

print(location)
print(date)
print(hour)
print(temp)
print(weather_code)

data_row = [location, dt, temp, weather_code, precip_bool]
data_row
```

```python
if exists('data/met_past_24h.csv'):
    with open('data/met_past_24h.csv', 'a') as f:
        f.write(str(data_row)[1:-1] + '\n')
else:
    with open('data/met_past_24h.csv', 'a') as f:
        f.write('location, date, hour, temp, weather_code, was_raining' + '\n')
        f.write(str(data_row)[1:-1] + '\n')
```

MariaDB met_past_24h table creation:

CREATE TABLE met_past_24h(
    location VARCHAR(255),
    date_time DATETIME,
    temp FLOAT,
    weather_code SMALLINT,
    precip_bool BOOL,
    PRIMARY KEY(location, date_time)
);

AccuWeather json data extraction

```python
import json, os, pprint, datetime
from pipeline_funcs import read_last_line
```

```python
json_line = read_last_line('data/accu_5d.jsonl')
pprint.pp(json_line)
```

```python
pprint.pp(json_line['Headline'])

date0_split = str.split(json_line['Headline']['EffectiveDate'][:10], '-')
date0_split = [int(x) for x in date0_split]
date0 = datetime.date(date0_split[0], date0_split[1], date0_split[2])
```

```python
pprint.pp(json_line['DailyForecasts'][0])
location = 'CREDENHILL'
return_list = []
for d in json_line['DailyForecasts']:
    forecast_date_str = d['Date'][:10]
    forecast_date_split = str.split(d['Date'][:10], '-')
    forecast_date_split = [int(x) for x in forecast_date_split]
    forecast_date = datetime.date(forecast_date_split[0], forecast_date_split[1], forecast_date_split[2])
    days_delta = str(forecast_date - date0)[:1]

    temp_max = (d['Temperature']['Maximum']['Value'] -25) * 5 / 9
    temp_min = (d['Temperature']['Minimum']['Value'] -25) * 5 / 9
    precip_bool = d['Day']['HasPrecipitation']
    return_list.append({'location' : location, 'date' : forecast_date, 'forecast_delta' : days_delta,
                        'temp_max' : temp_max, 'temp_min' : temp_min, 'precip_bool' : precip_bool})
    
    
```
